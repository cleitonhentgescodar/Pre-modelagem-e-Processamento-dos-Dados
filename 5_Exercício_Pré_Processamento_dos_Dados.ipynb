```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregar o data frame
df = pd.read_csv('/content/CHURN_TELECON_TAREFA .csv' , delimiter=';')

df.head(10)

# Visualização do data frame
```

```python
# Verificando tipos de dados
df.dtypes
```

```python
#Este código retornará True se houver algum valor nulo em qualquer lugar do DataFrame, e False caso contrário.
df.isnull().values.any()
```

```python
# Removendo e Tratando dados nulos

valores_nulos = df.isnull().sum()
print(valores_nulos)

porcentagem_nulos = (valores_nulos / len(df)) * 100
print(porcentagem_nulos)

df['PhoneService'] = df['PhoneService'].fillna('no')

df['Pagamento_Mensal'] = pd.to_numeric(df['Pagamento_Mensal'], errors='coerce')

media_pagamento_mensal = df['Pagamento_Mensal'].mean()
#print([df['Pagamento_Mensal'].mean()])

df['Pagamento_Mensal'] = df['Pagamento_Mensal'].fillna(media_pagamento_mensal)

df['Churn'] = df['Churn'].fillna('Yes')

# Como não é indicado definir a variável Genero pois altera diretamente nosso publico optei por deletar linhas faltantes de sexo masculino ou feminino.
df = df.dropna(subset=['Genero'])

valores_nulos = df.isnull().sum()
#print(valores_nulos)

porcentagem_nulos = (valores_nulos / len(df)) * 100
print('Revisão:\n', porcentagem_nulos)
```

```python
# Filtrar dados escritos errados

df.loc[:, 'Genero'] = df['Genero'].replace(['f', 'F'], 'Female')
df.loc[:, 'Genero'] = df['Genero'].replace(['m', 'M'], 'Male')
print(df['Genero'].unique())

df.loc[:, 'Casado'] = df['Casado'].replace(['yes', 'y'], 'Yes')
df.loc[:, 'Casado'] = df['Casado'].replace(['no', 'n'], 'No')
print(df['Casado'].unique())

df.loc[:, 'Dependents'] = df['Dependents'].replace(['yes', 'y'], 'Yes')
df.loc[:, 'Dependents'] = df['Dependents'].replace(['no', 'n'], 'No')
print(df['Dependents'].unique())

df.loc[:, 'PhoneService'] = df['PhoneService'].replace(['yes', 'y'], 'Yes')
df.loc[:, 'PhoneService'] = df['PhoneService'].replace(['no', 'n'], 'No')
print(df['PhoneService'].unique())

df.loc[:, 'Servico_Internet'] = df['Servico_Internet'].replace(['dsl'], 'DSL')
print(df['Servico_Internet'].unique())

df.loc[:, 'Servico_Seguranca'] = df['Servico_Seguranca'].replace(['No internet service'], 'No')
print(df['Servico_Seguranca'].unique())

df.loc[:, 'Suporte_Tecnico'] = df['Suporte_Tecnico'].replace(['No internet service'], 'No')
print(df['Servico_Seguranca'].unique())

df.loc[:, 'StreamingTV'] = df['StreamingTV'].replace(['No internet service'], 'No')
print(df['StreamingTV'].unique())
```

```python
# Valores estatísticos do Pagamento_Mensal

media = df['Pagamento_Mensal'].mean()
print('media:',media)

mediana = df['Pagamento_Mensal'].median()
print('mediana:',mediana)

moda = df['Pagamento_Mensal'].mode()
print('moda:',moda)

df['Pagamento_Mensal'].describe()
```
